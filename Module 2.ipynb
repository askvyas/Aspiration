```python
#Task6
#unfiltered dataframes
itc_uf=pd.read_csv('Large_Cap/ITC.csv')
cipla_uf=pd.read_csv('Large_Cap/CIPLA.csv')
reddy_uf=pd.read_csv('Large_Cap/DRREDDY.csv')
hdfc_uf=pd.read_csv('Large_Cap/HDFC.csv')
lt_uf=pd.read_csv('Large_Cap/LT.csv')
```

```python
itc_f=itc_uf[itc_uf.Series=="EQ"]
cipla_f=cipla_uf[cipla_uf.Series=="EQ"]
reddy_f=reddy_uf[reddy_uf.Series=="EQ"]
hdfc_f=hdfc_uf[hdfc_uf.Series=="EQ"]
lt_f=lt_uf[lt_uf.Series=="EQ"]
```

```python
columns = ['ITC','CIPLA','DRREDDY','HDFC','LT']
close_prices_dataFrame = pd.DataFrame(columns = columns)
close_prices_dataFrame['ITC'] = itc_f['Close Price']
close_prices_dataFrame['CIPLA'] = cipla_f['Close Price']
close_prices_dataFrame['DRREDDY'] = reddy_f['Close Price']
close_prices_dataFrame['HDFC'] = hdfc_f['Close Price']
close_prices_dataFrame['LT'] = lt_f['Close Price']
close_prices_dataFrame.dropna()
```

```python
pct_change_dataFrame = close_prices_dataFrame.pct_change().fillna(0)
pct_change_dataFrame.dropna()
```

```python
sns.pairplot(pct_change_dataFrame)
```

```python
#Task7
rolling_avg_lt = pct_change_dataFrame['LT'].rolling(7).mean()
rolling_avg_lt
```

```python
stand_lt = rolling_avg_lt.fillna(0).std()
stand_lt
```

```python
crrDate = pd.to_datetime(lt_f['Date'])

plt.figure(figsize=(20,10))
plt.plot(crrDate.tolist(),rolling_avg_lt.fillna(0).tolist())
```
