```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
from sklearn.neural_network import MLPClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

from sklearn.ensemble import RandomForestClassifier
```

```python
lt_bd=pd.read_csv('LT_bol.csv').dropna().reset_index()
lt_bd
```

```python
lt_bd['upper'] = lt_bd['upper'].fillna(0)
lt_bd['lower'] = lt_bd['lower'].fillna(0)
lt_bd['mid']=(lt_bd['lower']+lt_bd['upper'])/2
```

```python
lt_bd
```

```python
def calc_c(lt_bd):
    if (lt_bd['LT']<lt_bd['lower']):
        return "Buy"
    elif ((lt_bd['LT']>lt_bd['lower'])&(lt_bd['LT']<lt_bd['mid'])):
        return "Hold Buy/Liquidate Short"
    elif ((lt_bd['LT']>lt_bd['mid'])&(lt_bd['LT']<lt_bd['upper'])):
        return "Hold Short/Liquidate Buy"
    elif (lt_bd['LT']>lt_bd['upper']):
        return "Short"
        
    
```

```python
lt_bd = lt_bd.assign(Call = lt_bd.apply(calc_c, axis=1))
```

```python
lt_bd
```

```python
columns = ['lower', 'upper', 'mid']
tmp=lt_bd[columns]
```

```python
X=tmp.to_numpy()
```

```python
le = preprocessing.LabelEncoder()

label_Y = le.fit_transform(lt_bd[['Call']])
Y = label_Y.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=27)
```

```python
lda = LDA(n_components=1)
X_train = lda.fit_transform(X_train, y_train)
X_test = lda.transform(X_test)
```

```python
#Logistic Regression
log_ref= LogisticRegression()
log_ref.fit(X_train,y_train)
pred=log_ref.predict(X_test)
print(accuracy_score(pred, y_test))
```

```python
#KNN
KNN_model = KNeighborsClassifier(n_neighbors=5)
KNN_model.fit(X_train, y_train)
pred_2=KNN_model.predict(X_test)
print(accuracy_score(pred_2,y_test))
```

```python
#RandomForest
clf=RandomForestClassifier(n_estimators=100)
clf.fit(X_train,y_train)
pred_3=clf.predict(X_test)
print(accuracy_score(pred_3,y_test))
```

```python
#neural nets

NN = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)  
NN.fit(X_train, y_train)  
NN.predict(X_train)
print("Neural Networks")
round(NN.score(X_train, y_train), 4)
```

```python
tcs_dat = pd.read_csv('Large_Cap/TCS.csv')
tcs_dat['14 Day MA'] = tcs_dat['Close Price'].rolling(window=14).mean()
tcs_dat['30 Day STD'] = tcs_dat['Close Price'].rolling(window=20).std()
tcs_dat['upper'] = tcs_dat['14 Day MA'] + (tcs_dat['30 Day STD'] * 2)
tcs_dat['lower'] = tcs_dat['14 Day MA'] - (tcs_dat['30 Day STD'] * 2)
tcs_dat = tcs_dat.dropna()
tcs_dat['mid'] = (tcs_dat['upper']+tcs_dat['lower'])/2
tcs_dat
```

```python
def calc_c(tcs_dat):
    if (tcs_dat['Close Price']<tcs_dat['lower']):
        return "Buy"
    elif ((tcs_dat['Close Price']>tcs_dat['lower'])&(tcs_dat['Close Price']<tcs_dat['mid'])):
        return "Hold Buy/Liquidate Short"
    elif ((tcs_dat['Close Price']>tcs_dat['mid'])&(tcs_dat['Close Price']<tcs_dat['upper'])):
        return "Hold Short/Liquidate Buy"
    elif (tcs_dat['Close Price']>tcs_dat['upper']):
        return "Short"
        
    
```

```python
tcs_dat = tcs_dat.assign(Call = tcs_dat.apply(calc_c, axis=1))
```
