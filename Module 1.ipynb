```python
#Task4
df['Month'] = pd.DatetimeIndex(df['Date']).month
df['Year'] = pd.DatetimeIndex(df['Date']).year
df['VWAP'] = (df['Close Price'] * df['Total Traded Quantity']).cumsum() / df['Total Traded Quantity'].fillna(0).cumsum()
data_vwap = df[['Month','Year','VWAP']]
group = data_vwap.groupby(['Month','Year'])
group.first()
```

```python
#Task5
#part 1 Definition
def avg_n(n):
    return (df['Average Price'].tail(n).sum())/n
#part2 Defenition
def pl_perc(N):
    diff=(df['Close Price'].tail(N).iloc[N-1]-df['Close Price'].tail(N).iloc[0])
    if(diff<0):
        loss=-(diff)
        loss_perc=(loss/df['Close Price'].tail(N).iloc[N-1])*100
        return loss_perc
    if(diff>0):
        profit=diff
        profit_perc=(profit/df['Close Price'].tail(N).iloc[N-1])*100
        return profit_perc

#the averages
print("The average Values for 1 week,2 weeks,1 month,2 months,6 months and 1 year")
print(avg_n(5))
print(avg_n(10))
print(avg_n(20))
print(avg_n(60))
print(avg_n(120))
print(avg_n(240))

#Profit Loss
print("The profit loss percentage Values for 1 week,2 weeks,1 month,2 months,6 months and 1 year")

print(pl_perc(5))
print(pl_perc(10))
print(pl_perc(20))
print(pl_perc(60))
print(pl_perc(120))
print(pl_perc(240))


```

```python
#Task 6
df['Day_Perc_Change']=df['Close Price'].pct_change()
df['Day_Perc_Change'][0]=0
df
```

```python
#Task7
for i in range(493):
    if((df['Day_Perc_Change'][i]>=-0.5)&(df['Day_Perc_Change'][i]<=0.5)):
        df['Trend']="Slight or No Change"
    elif((df['Day_Perc_Change'][i]>=0.5)&(df['Day_Perc_Change'][i]<=1.0)):
        df['Trend']="Slight Positive"
    elif((df['Day_Perc_Change'][i]<=-0.5)&(df['Day_Perc_Change'][i]>=-1.0)):
        df['Trend']="Slight Negative"
    elif((df['Day_Perc_Change'][i]>=1)&(df['Day_Perc_Change'][i]<=3.0)):
        df['Trend']="Positive"
    elif((df['Day_Perc_Change'][i]>=-3)&(df['Day_Perc_Change'][i]<=-1.0)):
        df['Trend']="Negative"
    elif((df['Day_Perc_Change'][i]<=3)&(df['Day_Perc_Change'][i]>=7.0)):
        df['Trend']="Among Top Gainers"
    elif((df['Day_Perc_Change'][i]>=-7.0)&(df['Day_Perc_Change'][i]<=-3.0)):
        df['Trend']="Among Top Gainers"
    elif(df['Day_Perc_Change'][i]>=7.0):
        df['Trend']="Ball Run"
    elif(df['Day_Perc_Change'][i]<=-7.0):
        df['Trend']="Bear Drop"
    
    
        
        
df
        
        
        
```
